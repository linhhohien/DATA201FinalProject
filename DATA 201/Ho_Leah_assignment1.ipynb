## 6. Identify and define the first four moments (mean, variance, skewness, kurtosis) of the log return distribution.

```python
print('Mean = ' + str(info['natural_log'].mean()))
```

Mean: The mean is 0.0167 -> the average log return is 1.67% per period. This points to a long-term, positive predicted return.

```python
print('Standard Deviation = ' + str(info['natural_log'].std()))
```

Variance:
- The standard deviation is 0.0609 -> log returns deviate from the mean by approximately 6.09%.
- The variance is the square of the standard deviation = (0.0609)^2 = 0.00371 
- This quantifies the volatility of returns.

```python
from scipy.stats import skew
skewness = skew(info['natural_log'].dropna())
print('Skew = ' + str(skewness))
```

Skewness: Negative -> the distribution has a slightly longer left tail, meaning there are slightly more extreme 
negative returns than positive ones.

```python
from scipy.stats import kurtosis
kurtosis = kurtosis(info['natural_log'].dropna())
print('Kurtosis = ' + str(kurtosis))
```

Kurtosis: Close to zero -> the distribution has light tails (less extreme price movements than a normal distribution).

## 8. Exploring Normality: Research the Jarque-Bera (JB) test and its application to financial data. Perform the test using scipy.stats.jarque bera(), and interpret the results in the context of stock return distributions. Discuss why normality is important (or not) in financial modeling.

```python
from scipy.stats import jarque_bera
jb_stat, jb_p_value = jarque_bera(info['natural_log'].dropna())
print(f"Jarque-Bera Statistic = {jb_stat}")
print(f"P-Value = {jb_p_value}")
```

The Jarque-Bera (JB) test is a statistical test used to determine whether a dataset follows a normal distribution. 
It is particularly useful in financial modeling, where the assumption of normality impacts risk assessment and forecasting models.

**Application in Financial Data**: Stock returns are often assumed to be normally distributed in theoretical models, but real-world data frequently exhibit skewness and kurtosis, leading to deviations from normality.

**Interpretation of Results**:
  - If the p-value is **less than 0.05**, we reject the null hypothesis, indicating that the log returns do not follow a normal distribution.
  - If the p-value is **greater than 0.05**, we fail to reject the null hypothesis, suggesting that the log returns might be normally distributed.

**Importance of Normality in Financial Modeling:**
- Many financial models, such as the Black-Scholes option pricing model and Value-at-Risk (VaR) calculations, assume normality of returns.
- If returns are not normally distributed, risk can be underestimated, leading to mispricing of assets and inaccurate risk measures.
- Heavy tails (excess kurtosis) in financial returns suggest higher probabilities of extreme losses, which must be accounted for in risk management.

**Conclusion:** Based on the Jarque-Bera test results, if our log returns deviate significantly from normality, alternative models such as fat-tailed distributions or non-parametric approaches should be considered for more accurate financial analysis.
